```python
Y_train = titanic_train["Survived"]
X_train = titanic_train.drop(["Survived"], axis=1)
print(X_train)
```

```python
#converting pandas dataframe to arrays
Y_train = np.array(Y_train)
X_train = np.array(X_train)
print(X_train)
print(Y_train.shape)
```

```python
#logistic regression algorithm
def algorithm(X_train, Y_train):
    learning_rate = 0.18
    iterations = 1300
    theta = np.zeros(12)
   
    
    for i in range(iterations):
        deno = np.exp(-np.matmul(X_train, theta))
        
        hypothesis = 1/(1 + deno)
        
        J = hypothesis - Y_train
        
        theta = theta -(learning_rate * np.matmul(X_train.T, J))/ X_train.shape[0]
    
    return theta, hypothesis
   
  
```

```python
theta, hypothesis = algorithm(X_train, Y_train)
print(theta)
```

```python
#function to return prediction
def prediction(hypothesis):
    Y_prediction = hypothesis
    Y_prediction[Y_prediction>=0.5] = 1
    Y_prediction[Y_prediction<0.5] = 0
        
    return Y_prediction
    
    
```

```python
#accuracy algorithm
Y_prediction = prediction(hypothesis)
def accuracy(Y_prediction, Y_train):
    c = np.count_nonzero(Y_prediction == Y_train)
    accuracy = (c/len(Y_train))*100
    return accuracy

accuracy = accuracy(Y_prediction, Y_train)
print (accuracy)
  
        
        
```
