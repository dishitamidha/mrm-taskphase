```python
#uploading dataset
import numpy as np
import pandas as pd

dataset = pd.read_csv(r'C:\Users\dishi\OneDrive\Desktop\task 1\project_1_dataset.csv')


dataset.head(5)
```

```python
#visually representing the dataset
from matplotlib import pyplot
pyplot.scatter(dataset['YearsExperience'],dataset['Salary'])
pyplot.title('Dataset Representation')
pyplot.ylabel('Salary')
pyplot.xlabel('Years Experience')
pyplot.show()
```

```python
#splitting dataset into testing and training sets
from sklearn.model_selection import train_test_split
X = dataset['YearsExperience']
y = dataset['Salary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)
X=np.array(X)
y=np.array(y)
X_train=np.array(X_train)
X_test=np.array(X_test)
y_train=np.array(y_train)
y_test=np.array(y_test)

```

```python
#algorithm for linear regression

def gradient_descent(X_train, y_train):
    learning_rate= 0.000010
    iterations= 2800
    theta=np.zeros(2)
    ones=np.ones(len(X_train))
    b= np.column_stack((ones,X_train))

    hypothesis = np.dot(b,theta.T) 
    
    
    J = (hypothesis - y_train)

    
    for i in range(iterations):
        theta[0] = theta[0] - (learning_rate * sum(J))/len(X_train)
        theta[1] = theta[1] - (learning_rate * sum(J*X_train))/len(X_train)
    y_train_prediction = theta[0] + theta[1]*X_train

    return y_train_prediction, theta
  
a,d = gradient_descent(X_train, y_train)
print(d)   
```

```python
#visualising training set
pyplot.scatter(X_train, y_train)
pyplot.title('Training set reprensentation')
pyplot.ylabel('Salary')
pyplot.xlabel('Years Experience')
pyplot.show()



#visualising linear regression in training set
pyplot.scatter(X_train, y_train)
pyplot.plot(X_train, a, color='green')
pyplot.title('Training set reprensentation')
pyplot.ylabel('Salary')
pyplot.xlabel('Years Experience')
pyplot.show()


```

```python
#visualising testing set
pyplot.scatter(X_test, y_test)
pyplot.title('Testing set reprensentation')
pyplot.ylabel('Salary')
pyplot.xlabel('Years Experience')
pyplot.show()

p,q = gradient_descent(X_test, y_test)

#visualising linear regression in testing set
pyplot.scatter(X_test, y_test)
pyplot.plot(X_test, p, color='green')
pyplot.title('Testing set reprensentation')
pyplot.ylabel('Salary')
pyplot.xlabel('Years Experience')
pyplot.show()


   
```
